#Natural Language Processing

* The transformation from words to vectors is also known as word embedding. The reason for this transformation is so that machine learning algorithm can perform linear algebra operations on numbers (in vectors) instead of words

* Việc chuyển đổi các từ về dạng các vectors được biết như là Word Embedding. Lý do là để machine có thể học được chúng bằng các phếp toán đại số tuyến tính.



# Pre-processing text data

* Large Movie Review Dataset.
* This is a dataset for binary sentiment classification containing substantially more data than previous benchmark datasets. 
* We provide a set of 25,000 highly polar movie reviews for training, and 25,000 for testing. 


* Dữ liệu là Large Movie Review Dataset của thư viện tensorflows có sẵn.
* Là binary class gồm: negative và positive

* https://www.tensorflow.org/datasets/catalog/imdb_reviews


```
#load dữ liệu
import tensorflow
import tensorflow_datasets as tfds

imdb, info = tfds.load('imdb_reviews', with_info = True, as_supervised = True)
```

```
#imdb gồm train và test

train_data, test_data = imdb['train'], imdb['test']
```

```
#tách các reviews và labels từ dữ liệu train và test, sau đó tạo corpus.

train_sequences = []
test_sequences = []

train_labels = []
test_labels = []

for sentences, labels in train_data:
  train_sequences.append(sentences.numpy().decode("utf-8"))
  train_labels.append(labels.numpy())
for sentences, labels in test_data:  
  test_sequences.append(sentences.numpy().decode("utf-8"))
  test_labels.append(labels.numpy())
```

```
import re 

def normalize_text(text):

    #Remove các ký tự kéo dài: vd: đẹppppppp
    text = text.lower()

    text = re.sub(r"[^A-Za-z]+"," ", text)

    # Chuyển thành chữ thường
    return text
```
